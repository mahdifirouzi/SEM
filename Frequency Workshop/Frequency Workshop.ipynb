- Here is the whole code

```python
# Here is the whole code
# Import Pandas Library
import pandas as pd

# Assuming your CSV file is named 'f 2019 8.csv' may be you have to change it to your google colab path
file_path = 'fnew-2023-12.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path, sep=',', parse_dates=['dtm'])

# Filter out rows where 'f' is zero
non_zero_df = df[df['f'] != 0]

# Find the row with the minimum non-zero frequency
min_non_zero_row = non_zero_df.loc[non_zero_df['f'].idxmin()]

# Extract timestamp and value
min_timestamp = min_non_zero_row['dtm']
min_value = min_non_zero_row['f']

# Print the result
print("Minimum non-zero frequency timestamp and value:")
print(f"Timestamp: {min_timestamp}")
print(f"Value: {min_value}")
```

- Let plot The Frequency before and after the incident

```python
import matplotlib.pyplot as plt
# Plot a line chart from half an hour before to half an hour after the incident
half_hour_before = min_timestamp - pd.Timedelta(minutes=30)
half_hour_after = min_timestamp + pd.Timedelta(minutes=30)

incident_df = df[(df['dtm'] >= half_hour_before) & (df['dtm'] <= half_hour_after)]

plt.figure(figsize=(40, 10))
plt.plot(incident_df['dtm'], incident_df['f'], linestyle='-', color='g')
plt.title('Frequency Around Incident')
plt.xlabel('Timestamp')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Plot a line chart from one minutes before and after the incident
half_hour_before = min_timestamp - pd.Timedelta(seconds=120)
half_hour_after = min_timestamp + pd.Timedelta(seconds=120)

incident_df = df[(df['dtm'] >= half_hour_before) & (df['dtm'] <= half_hour_after)]

plt.figure(figsize=(40, 10))
plt.plot(incident_df['dtm'], incident_df['f'], linestyle='-', color='g')
plt.title('Frequency Around Incident')
plt.xlabel('Timestamp')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
